```python
import nltk
# nltk.download()
# Importa o "tokenizador regex"
from nltk.tokenize import RegexpTokenizer
# Lista de StopWords em português
stopwords = nltk.corpus.stopwords.words('portuguese')
# REGEX que "filtra" somente palavras e números
tokenizer = RegexpTokenizer(r'\w+')
# Abre um texto salvo no Drive (substituir o caminho)
arquivo = open("/content/drive/MyDrive/!!!!!!!!!!PLN/Rio_Grande.txt", "r")
# Armazena a "leitura do arquivo"
texto = arquivo.read()
# Fecha o arquivo (conteúdo continua na variavel texto)
arquivo.close()

# Usando o regex para tokenizar nosso texto
tokens = tokenizer.tokenize(texto)

# -- Coloca as palavras em minusculo e verifica se as mesmas não são stopwords
lista_minuscula = []
# Faz um loop de repetição para cada token
for token in tokens:
  # Verifica se o token não é uma stopword
  if token.lower() not in stopwords:
    # Adiciona o token a lista
    lista_minuscula.append(token.lower())

# Cria a distribuição de frequencia
frequencia = nltk.FreqDist(lista_minuscula)
# Imprime as 10 palavras mais comuns do texto
frequencia.most_common(10)
```

Aplicamos um valor default para estas palavras que ele não identifica

```python
# Corpus etiqueta
# nltk.download('mac_morpho')
# Importa o corpus para dentro do programa
from nltk.corpus import mac_morpho
# Etiqueta cada palavra do nosso texto, uni significa que ele olha uma a uma
from nltk.tag import UnigramTagger
# Pega os tokens do ac_morpho e joga dentro de nossa variavel sentecas_treino

# Importar o etiquetador default
from nltk import DefaultTagger
# Cria a tag padrão
tag_padrao = DefaultTagger('N')


sentencas_treino = mac_morpho.tagged_sents()

# Etquetador pega o material de estudo (sentencas_treino) e cria uma espécie de dicionario para etiquetar as nossas palavras
# Versão 2.0!!!
etiquetar = UnigramTagger(sentencas_treino, backoff=tag_padrao)

# Classifica as palavras do nosso texto
palavras_etiquetadas = etiquetar.tag(tokens)
palavras_etiquetadas
#
```
