Stemming

Técnica de PLN que consiste em reduzir as palavras removendo sufixos e prefixos

Agrupa palavras com significados semelhantes sob uma mesma forma (correr, correndo e corrido vão se tornar corr)

para que isso? Facilita o processamento de texto

Usaremos para isso um objeto da classe RSLPStemmer


```python
stemmer = nltk.stem.RSLPStemmer()
```

RSLPStemmer é uma classe dentro do nltk.stem que possui um algoritmo de stemming para a língua portuguesa

Dentro do algoritmo existem regras que vão seguir caracteristicas morfológicas do portugues para tentar reduzir as palavras de forma precisa

Voltando ao DataFrame, criaremos uma coluna chamada Stemmer. Esta coluna vai receber o resultado do stemmer

Como é palavra por palavra temos que iterar sobre cada palavra da nossa coluna Sem processamento.

```python
palavras['Stemmer'] = [stemmer.stem(palavra) for palavra in palavras['Sem processamento']]
```

imprimir o resultado

```python
palavras
```

**Lematização**

A lematização é um processo de PLN que visa reduzir uma palavra as sua forma base, conhecida como lema. A ideia seria parecida com o Stemming só que com algumas caracteristicas diferentes

* Stemming remove sufixos e prefixos mas muitas vezes não retorna uma palavra existente (corr etc)
* Lematização leva em consideração o significado da palavra e o contexto, tentando chegar a um lema valido - > correndo = correr

Necessario a utilização de um dicionário lexical

No nosso caso baixamos tudo do nltk, poderiamos baixar só o dicionario com o comando a seguir

```python
nltk.download('wordnet')
```

Vamos usar o nosso dicionario

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

Criando o lematizador

```python
lematizador = WordNetLemmatizer()
```

```python
palavra = "running"

lema = lematizador.lemmatize(palavra)
print(lema)
```

Aprimorando o lemaizador indicando a classe gramatical

```python
palavra = "running"

lema = lematizador.lemmatize(palavra, pos=wordnet.VERB)
print(lema)
```
