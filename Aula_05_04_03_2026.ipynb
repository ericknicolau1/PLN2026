Tokenização -> transformar um texto em pedaços menores, estes pedaços menores são os tokens

Tokens - > palavra individuais, pontuações, números, etc

A ideia de ideia de tokenizar algo é transformar o texto em uma estrutra que o computador consiga analisar/processar

Primeira parte, tokenização de palavras (word-tokenize)

```python
arquivo = open("/content/drive/MyDrive/!!!!!!!!!!PLN/Rio_Grande.txt", "r")
texto = arquivo.read()
arquivo.close()
tokens = nltk.word_tokenize(texto)
print(tokens)
```

Frequência de Palavras

```python
arquivo = open("/content/drive/MyDrive/!!!!!!!!!!PLN/Rio_Grande.txt", "r")
texto = arquivo.read()
arquivo.close()
tokens = nltk.word_tokenize(texto)
frequencia = nltk.FreqDist(tokens)
frequencia.most_common(10)
```

Vamos usar Regex para eliminar a pontuação

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
arquivo = open("/content/drive/MyDrive/!!!!!!!!!!PLN/Rio_Grande.txt", "r")
texto = arquivo.read()
arquivo.close()
tokens = tokenizer.tokenize(texto)
frequencia = nltk.FreqDist(tokens)
frequencia.most_common(10)
```

Colocando todo o texto em minúsculas para verificar as palavras mais comuns

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
arquivo = open("/content/drive/MyDrive/!!!!!!!!!!PLN/Rio_Grande.txt", "r")
texto = arquivo.read()
arquivo.close()
tokens = tokenizer.tokenize(texto)

lista_minuscula = []
for token in tokens:
  lista_minuscula.append(token.lower())

frequencia = nltk.FreqDist(lista_minuscula)
frequencia.most_common(20)
```

StopWords - > são palavras sem grande valor para o processamento de sentido de um texto

```python
# Exemplo de StopWords
stopwords = nltk.corpus.stopwords.words('portuguese')
print(stopwords)
```

```python
# Importa o "tokenizador regex"
from nltk.tokenize import RegexpTokenizer
# Lista de StopWords em português
stopwords = nltk.corpus.stopwords.words('portuguese')
# REGEX que "filtra" somente palavras e números
tokenizer = RegexpTokenizer(r'\w+')
# Abre um texto salvo no Drive (substituir o caminho)
arquivo = open("/content/drive/MyDrive/!!!!!!!!!!PLN/Rio_Grande.txt", "r")
# Armazena a "leitura do arquivo"
texto = arquivo.read()
# Fecha o arquivo (conteúdo continua na variavel texto)
arquivo.close()

# Usando o regex para tokenizar nosso texto
tokens = tokenizer.tokenize(texto)

# -- Coloca as palavras em minusculo e verifica se as mesmas não são stopwords
lista_minuscula = []
# Faz um loop de repetição para cada token
for token in tokens:
  # Verifica se o token não é uma stopword
  if token.lower() not in stopwords:
    # Adiciona o token a lista
    lista_minuscula.append(token.lower())

# Cria a distribuição de frequencia
frequencia = nltk.FreqDist(lista_minuscula)
# Imprime as 10 palavras mais comuns do texto
frequencia.most_common(10)
```
