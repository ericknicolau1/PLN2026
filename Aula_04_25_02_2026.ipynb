Arquivos

É tipo nativo do Python - manipula arquivos sem precisar de algum módulo

Objeto para representar o arquivo

r = read, leitura

w = write, escrita

a = append, escrever no final do arquivo

terminou de usar um arquivo, fechar o arquivo, close()

```python
arquivo = open('/content/drive/MyDrive/!!!!!!!!!!PLN/Rio_Grande.txt', 'r')
```

```python
print(arquivo.read())
```

Criando um arquivo

```python
arquivo_escrita = open('/content/drive/MyDrive/!!!!!!!!!!PLN/arquivo_escrita2026.txt', 'w')
arquivo_escrita.write('Primeira linha\n')
arquivo_escrita.close()
```

Lendo o arquivo Criado

```python
arquivo_leitura = open('/content/drive/MyDrive/!!!!!!!!!!PLN/arquivo_escrita2026.txt', 'r')
print(arquivo_leitura.read())
arquivo_leitura.close()
```

Adicionando conteúdo a um arquivo existente

```python
arquivo_adicao= open('/content/drive/MyDrive/!!!!!!!!!!PLN/arquivo_escrita2026.txt', 'a')
arquivo_adicao.write('Segunda linha\n')
arquivo_adicao.close()
```

len() -> tamanho da string (caracteres e espaços)

```python
teste= "Hoje é Quarta-feira"
print(teste)
len(teste)
```

Colocar frase em minusculo

```python
print(teste.lower())
```

Colocar em maiusculo

```python
print(teste.upper())
```

Vamos ajustar uma frase removendo espaços antes e depois da string

```python
teste2= "                  Hoje é domingo                    "
len(teste2)
```

```python
ajustado = teste2.strip()
print(ajustado)
len(ajustado)
```

Trocar conteúdo

```python
trocando = ajustado.replace('domingo', 'quarta-feira')
print(trocando)
```

```python
trocando = ajustado.replace('sexta', 'quarta-feira')
print(trocando)
```

split() -> divide a string

```python
teste3 = "Hoje é quarta-feira, dia 25 de fevereiro de 2026, o dia esta nublado"
fatiado = teste3.split()
print(fatiado)
```

Escolher onde vai dividir

```python
fatiado2 = teste3.split(',')
print(fatiado2)
```

```python
fatiado3 = teste3.split('e')
print(fatiado3)
```

Novas funções através de combinações

```python
teste4 = "Érick Henrique Pereira Nicolau"
print(len(teste4))
print(len(teste4.split()))
print(teste4.split())
```

Dividindo e juntando

```python
teste5 = "Érick Henrique Pereira Nicolau"
fatiando = teste5.split()
print(fatiando)
```

```python
juntando = " ".join(fatiando)
print (juntando)
```

```python
juntando = "->".join(fatiando)
print (juntando)
```
